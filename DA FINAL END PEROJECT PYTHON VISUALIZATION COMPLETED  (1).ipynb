```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Distribution of Customer Rating
sns.histplot(df['Rating'], bins=10)
plt.title("Distribution of Customer Ratings")
plt.show()
```

```python
#Product Line Count
df['Product line'].value_counts()
```

```python
#BIVARIATE ANALYSIS
```

```python
 # Total Sales by Gender
df.groupby('Gender')['Total'].sum()
```

```python
sns.barplot(x='Gender', y='Total', data=df)
plt.title("Total Sales by Gender")
plt.show()
```

```python
#Payement Method distribution
sns.countplot(x='Payment', data=df)
plt.title("Payment Methods Used")
plt.show()
```

```python
#MULTIVARIATE ANALYSIS
```

```python
#ANALYSIS OF PRODUCT LINE SALES IN  EACH CITY 
pivot = pd.pivot_table(df,
                       values='Total',
                       index='City',
                       columns='Product line',
                       aggfunc='sum')

pivot
```

```python
#CORRELATION HEATMAP
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#TOTAL SALES BY PRODUCT LINE USING A BAR CHART 
df = pd.read_csv("WalmartSalesData.csv (1).csv")

sales_product = df.groupby('Product line')['Total'].sum()

sales_product.plot(kind='bar')
plt.title("Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Daily Sales Trend USING A LINE CHART
daily_sales = df.groupby('Date')['Total'].sum()

daily_sales.plot()
plt.title("Daily Sales Trend")
plt.show()
```

```python
#Payment Method Distribution Using a PIE Chart
payment = df['Payment'].value_counts()

plt.pie(payment, labels=payment.index, autopct='%1.1f%%')
plt.title("Payment Method Distribution")
plt.show()
```

```python
#Sales Distribution by Product Line Uisng aBOX PLOT
sns.boxplot(x='Product line', y='Total', data=df)
plt.xticks(rotation=45)
plt.title("Sales Distribution by Product Line")
plt.show()
```

```python
#Analysis of Unit Price vs Total Sales
sns.scatterplot(x='Unit price', y='Total', data=df)
plt.title("Unit Price vs Total Sales")
plt.show()
```
